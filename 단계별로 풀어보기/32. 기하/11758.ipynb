# 11758 기하

```python
A = [0, 0, 0]
for i in range(3):
    A[i] = list(map(int, input().split()))
```

```python
line1 = (A[0][0]*A[1][1] + A[1][0]*A[2][1] + A[2][0]*A[0][1])
line2 = (A[1][0]*A[0][1] + A[2][0]*A[1][1] + A[0][0]*A[2][1])
result = line1 - line2
```

```python
if result > 0:
    print('1')
elif result < 0:
    print('-1')
else:
    print('0')
```
